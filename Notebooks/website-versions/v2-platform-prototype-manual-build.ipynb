# V2 Learning Notebook: Product Prototype

## Version
`v2` in `web/platform-prototype/`

## Why this notebook exists
This notebook explains how two humans could build the interactive product prototype manually and how to talk about that work in the presentation.

## What this version is for

This version is the product-concept layer.

It is useful when you want to show:
- how the product feels to use
- what a planning workflow looks like
- how metrics and scenarios appear together
- how the capstone can be presented as a real product

A good presentation sentence is:
"This prototype shows the interaction model of the product. It is not the production engine itself, but it demonstrates how a user would experience the planning and risk workflow."

## How two teammates could divide the work

### Teammate A: product logic and scenario design
- Read the real app flow in `app.py`.
- Decide which modules belong in the prototype.
- Define realistic sample scenarios and KPI values.
- Make sure the prototype stays aligned with the real capstone logic.

### Teammate B: interaction design and frontend build
- Create the layout system.
- Build the navigation and card hierarchy.
- Implement scenario switching in JavaScript.
- Test responsiveness and visual clarity.

### Shared checkpoints
- Confirm the prototype does not misrepresent the real app.
- Review whether the metrics make sense together.
- Align the visual flow with the demo script.
- Practice the walkthrough as a team.


## Step-by-step human workflow

### 1. Read the real product flow first
Use:
- `app.py`
- `README.md`
- `presentation/demo-script.md`

Humans should understand the real logic before making a polished prototype.

### 2. Decide the core modules
The final prototype uses modules such as:
1. project brief composer
2. executive forecast
3. risk distribution
4. critical path view
5. scenario lab
6. evidence pack

### 3. Create the files
```bash
mkdir -p platform-prototype
touch web/platform-prototype/index.html web/platform-prototype/styles.css web/platform-prototype/main.js
```

### 4. Build the layout before the interactions
Humans should first place cards and sections, then wire dynamic behavior later.

### 5. Create believable mock data
This is a major human step. KPI values, delay probabilities, and scenario summaries should feel consistent and presentation-safe.

### 6. Add interaction logic
Typical JavaScript tasks:
- switch scenarios
- update KPI values
- update summary text
- swap recommendation states

### 7. Test as a demo, not just a page
The right question is not only "does it load?" but also "can we present from this without confusion?"

## Key design decisions and why they matter

- Use a dashboard-like layout.
  Why: this version should feel like a product, not a brochure.

- Keep the KPI hierarchy obvious.
  Why: presenters need to know where the audience should look first.

- Use static demo data rather than backend calls.
  Why: the prototype should be stable during presentations.

- Keep the scenario logic human-readable.
  Why: judges care about decisions, not only interface polish.

## How to explain this version in the presentation

Suggested explanation:

"We built this version as a high-fidelity product prototype. First we mapped the real app flow from the Streamlit product, then we selected the most presentation-relevant modules: inputs, executive forecast, risk distribution, critical path, and scenarios. One teammate focused on the product logic and believable scenario data, and the other translated that into a polished frontend. The goal was to show how the product would feel in use while staying faithful to the actual capstone functionality."

## Questions a coach may ask

### Why make a prototype if you already have the app?
Because prototypes let us present the interaction model and product story with more control and visual polish than the live app alone.

### How did you keep it truthful?
We used the real app structure and real forecasting concepts, then mocked only the presentation states and sample values.

### What did you build manually?
We manually designed the layout, chose the product modules, created the mock scenario data, implemented the interactions, and tested the demo flow ourselves.

## What you should learn from this version

- How to translate backend logic into a frontend concept.
- How to choose which features deserve visual emphasis.
- How to build a demo-first UI with stable mock states.
- How to explain prototype decisions during a capstone presentation.
