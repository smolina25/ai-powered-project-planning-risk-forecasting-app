# V3 Learning Notebook: Lovable-Style Multi-Page Site

## Version
`v3` in `web/lovable-style-site/`

## Why this notebook exists
This notebook teaches how two humans could manually recreate a Lovable-style multi-page site for the capstone and how to explain that process confidently in the presentation.

## What this version is for

This version is the polished marketing-style website.

Its job is to make the capstone feel like a product company website, not only a student repo.

It helps show:
- visual polish
- complete website navigation
- product positioning
- product-market style messaging
- a clean shareable surface for external viewers

A good presentation sentence is:
"This version was built to package the capstone as a complete product-facing website, using the structure and pacing of the Lovable reference while adapting the content to our actual project."

## How two teammates could divide the work

### Teammate A: reference analysis and content adaptation
- Inspect the reference Lovable site and video.
- Extract the repeating structure: nav, hero, sections, pricing, dashboard, footer.
- Decide what can be borrowed structurally and what must be rewritten truthfully.
- Write the actual page copy.

### Teammate B: frontend architecture and implementation
- Create the shared CSS system.
- Build the multi-page structure.
- Implement reusable nav and footer rendering in JavaScript.
- Test all routes and responsive behavior.

### Shared checkpoints
- Confirm the site is visually close to the reference.
- Confirm the claims are still honest to the capstone.
- Review the page flow as if you were an external visitor.
- Practice how to justify the design choices in the presentation.

## Step-by-step human workflow

### 1. Study the reference
Humans should start by observing:
- top navigation pattern
- hero composition and copy rhythm
- spacing and contrast
- dark background treatment
- section order
- pricing page structure
- dashboard teaser or product page logic

### 2. Define the page architecture
For this site, the route plan became:
1. `index.html`
2. `product.html`
3. `ai-forecasting.html`
4. `monte-carlo.html`
5. `pricing.html`
6. `dashboard.html`
7. `about.html`
8. `login.html`

### 3. Create the shared frontend system
A human team would decide early to centralize:
- colors
- typography
- button classes
- card shells
- responsive behavior
- header/footer logic

### 4. Build the home page first
The home page establishes the brand tone. If it fails, the rest of the site will feel weak.

### 5. Build the secondary pages
Recommended order:
- product
- AI forecasting
- Monte Carlo
- dashboard
- pricing
- about
- login

### 6. Fix all navigation links
A standalone site must behave like a real site. Every nav item and CTA should open a real page.

### 7. Test the site like an external user
A human QA pass should include:
- clicking every nav item
- checking hero alignment
- opening mobile layout
- testing the mock login flow
- making sure the site still feels coherent page to page

## Key design decisions and why they matter

- Match the structure, not the claims.
  Why: copying fabricated business proof would weaken credibility.

- Use one shared nav/footer renderer.
  Why: this reduces duplication and keeps the site consistent.

- Keep the dark cinematic visual language.
  Why: that is a major part of what makes the reference feel polished.

- Give every route a real page.
  Why: broken links immediately make a product site feel unfinished.

- Keep the content mapped to the capstone.
  Why: the site must still support your presentation and be defendable if questioned.

## Questions a coach may ask

### Did you just copy the reference?
No. We used the reference as a structural and visual model, but we rewrote the content and rebuilt the pages manually for our own product and capstone story.

### What did you actually implement yourselves?
We implemented the page architecture, the design system, the shared navigation/footer logic, the individual routes, and the responsive behavior ourselves.

### Why make a multi-page site?
Because it presents the capstone as a polished product offering and supports sharing beyond the presentation itself.
