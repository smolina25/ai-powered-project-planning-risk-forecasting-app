# V1 Learning Notebook: Case-Study Showcase

## Version
`v1` in `web/showcase/`

## Why this notebook exists
This notebook is meant to teach how two humans with sufficient know-how could build the showcase website themselves, step by step, and explain that process during the capstone presentation.

## What this version is for

This is not the product UI. It is the public case-study page.

Its job is to answer these questions for a coach or stakeholder:
- What problem did we solve?
- Why does it matter?
- What did we actually build?
- Where is the evidence?
- Why is this capstone credible?

So if you explain this version in the presentation, say:
"This version was designed as the public narrative layer. It packages the capstone into a clear story for coaches, judges, recruiters, or portfolio viewers."

## How two teammates could divide the work

### Teammate A: content and product framing
- Extract business questions from the capstone docs.
- Decide the page sections and story order.
- Write the actual copy in stakeholder language.
- Check whether the page satisfies the coach's communication requirements.

### Teammate B: frontend implementation
- Build the HTML skeleton.
- Create the CSS system.
- Wire the buttons and config-driven links in JavaScript.
- Test layout and responsiveness.

### Shared checkpoints
- Agree on the page structure before coding.
- Review whether the content is accurate and not exaggerated.
- Test that the page is readable on desktop and mobile.
- Validate that it helps the final presentation rather than distracting from it.

If asked in the presentation how the team worked, a good answer is:
"We split the work into narrative design and implementation, then reviewed accuracy together so the public page stayed aligned with the actual capstone evidence."

## Step-by-step human workflow

### 1. Gather the source material
Use the core capstone evidence as the content source:
- `docs/capstone-business-questions.md`
- `docs/technical-eda-summary.md`
- `docs/model-comparison.md`
- `presentation/certAIn-Project-Intelligence-Capstone.pptx`
- `README.md`

### 2. Decide the page structure
A human PM or product person would usually map the sections in this order:
1. Hero
2. Problem
3. Solution
4. Evidence
5. Deliverables
6. CTA

### 3. Build the file structure
Example commands:
```bash
mkdir -p showcase
touch web/showcase/index.html web/showcase/styles.css web/showcase/main.js web/showcase/config.js
```

### 4. Write the HTML first
Humans should not start with styling. First build a clean content skeleton.

### 5. Add the CSS system
Define:
- color palette
- typography
- spacing scale
- layout widths
- buttons and cards

### 6. Add JavaScript only for useful behavior
For this version, JS should stay lightweight. It mostly handles config-driven links and small page interactions.

### 7. Review the content for honesty
This matters in a capstone. Do not claim real deployments, users, or business outcomes unless you actually have them.

### 8. Test locally
```bash
python -m http.server 8000
```
Open:
`http://127.0.0.1:8000/web/showcase/`


## Key design decisions and why they matter

- Keep this page editorial, not dashboard-heavy.
  Why: the job here is storytelling, not product interaction.

- Link directly to evidence artifacts.
  Why: coaches want proof, not just good-looking copy.

- Use simple business language.
  Why: this page may be read by non-technical stakeholders.

- Separate config from layout.
  Why: a human team can swap repo/deck/demo URLs without rewriting the page.

One teammate focused on the narrative and information architecture, and the other implemented the site structure, styling, and dynamic links. ## Questions a coach may ask

### Why create this site if you already have slides?
Because the site works as a persistent public artifact. Slides are presentation-specific, but the showcase page is portfolio-friendly and easier to browse.

### Why not show the app directly here?
Because the audience for this version is broader than product users. The page is meant to explain the capstone, not replace the live demo.

### What did you actually do manually?
We manually defined the content structure, wrote the copy, built the HTML/CSS/JS files, connected the evidence links, and tested the page ourselves.

## What you should learn from this version

- How to turn technical work into a stakeholder-facing story.
- How to separate content structure from visual design.
- How to use a static microsite as capstone evidence.
- How to explain a communication artifact during a presentation.
